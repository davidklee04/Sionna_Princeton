```python
from ipyleaflet import Map, DrawControl, projections
from IPython.display import display

# Initialize the map centered at a default location
m = Map(center=(38, -101), zoom=4, scroll_wheel_zoom=True)

# Configure the drawing control to only allow rectangles
draw_control = DrawControl(
    rectangle= {
        "shapeOptions": {
            "color": "#0064ff",
            "fillOpacity": 0.1
        },
        "metric": ['km','m']
    },
    polygon={},
    circle={},
    circlemarker={},
    polyline={},
    marker={}
    
)


# Attach the callback to the draw control
m.add_control(draw_control)

# Display the map
display(m)
```

```python

if draw_control.last_draw['geometry'] is not None:
    coords = draw_control.last_draw['geometry']['coordinates'][0]
    for idx, (lng, lat) in enumerate(coords):
        print(f"Corner {idx}: lng={lng}, lat={lat}")
else:
    print("No rectangle drawed")
```
